```python
import pandas as pd

df = pd.read_csv('./data/wine.csv')
df.head()


```

```python

df = df.drop_duplicates()
```

```python
# Подсчет количества пропусков в каждой колонке
missing_data_count = df.isnull().sum()

# print("Количество пропущенных данных в каждой колонке:")
# print(missing_data_count)

# Подсчет количества пропусков в каждой колонке
total_rows =  len(df)
missing_data_percentage  = (missing_data_count / total_rows) * 100
# missing_data_percentage
# Удаление колонок с более чем 40% пропусков
threshold = 30
columns_to_drop = missing_data_percentage[missing_data_percentage > threshold].index
df_cleaned = df.drop(columns=columns_to_drop)
```

```python
df_cleaned

# Подсчет количества пропусков в каждой колонке
missing_data_count = df_cleaned.isnull().sum()
missing_data_count
```

```python
# все пропущенные страны заменили на unknown
df_cleaned['country'] = df_cleaned['country'].fillna('unknown')
df_cleaned['designation'] = df_cleaned['designation'].fillna('unknown')
df_cleaned['province'] = df_cleaned['province'].fillna('unknown')
df_cleaned['region_1'] = df_cleaned['region_1'].fillna('unknown')
df_cleaned['taster_name'] = df_cleaned['taster_name'].fillna('unknown')
df_cleaned['taster_twitter_handle'] = df_cleaned['taster_twitter_handle'].fillna('unknown')
df_cleaned['variety'] = df_cleaned['variety'].fillna('unknown')

# Заполнение пропусков уены медианным значением
median_price = df_cleaned['price'].median()
df_cleaned['price'] = df_cleaned['price'].fillna(median_price)

```
